## Checking valid covariance matrices

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt

# Define a function for checking the covariance of a matrix
def check(a):
    m=np.linalg.eigvals(a)
    if (np.transpose(a)==a).all() and np.all(m>=0):
        return print('The matrix is a valid covariance Matrix')
    else:
        return print('The matrix is not a valid covariance matrix')
```

```python
a=np.array([[-1,2],
            [3,4]])
check(a)
#The polynomial for above matrices is y=-x1²+4x2²+5x1x2
x1_vals=np.linspace(-10,10)
x2_vals=np.linspace(-10,10)
x1,x2=np.meshgrid(x1_vals,x2_vals)

y=-x1**2+4*x2**2+5*x1*x2

ax=plt.axes(projection='3d')
ax.plot_surface(x1,x2,y)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y');
```
