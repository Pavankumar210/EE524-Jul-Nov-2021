```python
#Import libraries
import matplotlib.pyplot as plt
import numpy as np

#Plotting The polynomial y
# Detemine x1 , x2 axes
x1_vals=np.linspace(-10,10)
x2_vals=np.linspace(-10,10)
x1,x2=np.meshgrid(x1_vals,x2_vals)

y=0.5*x1**2+0.5*x2**2

ax=plt.axes(projection='3d')
ax.plot_surface(x1,x2,y)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y');
```

```python
a=np.array([[0.5,0],
            [0,0.5]])
w,v=np.linalg.eig(a)
print('Eigen values are :\n',w)
print('Eigen vectors are:\n',v)
```
