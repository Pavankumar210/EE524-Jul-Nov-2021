# Uniform Distribution

## Probability Density Function

```python
import numpy as np
import matplotlib.pyplot as plt

#Define PDF in Uniform Distribution

def un_pdf(a,b):
    def fn(x):
        if a<x<b:
            return 1/(b-a)
        else:
            return 0
    return fn

#Take differnt values for a,b

f1=un_pdf(3,8) #a=3,b=8
f2=un_pdf(1,3) #a=1,b=3
f3=un_pdf(2,6) #a=2,b=6 
# plot functions for values of a taken

#graph1
X=np.linspace(0,10,20)
Y1=[f1(x) for x in X]
plt.plot(X,Y1,'r', label='a=3,b=8')

#graph2
Y2=[f2(x)for x in X]
plt.plot(X,Y2,'y', label='a=1,b=3')

#graph3
Y3=[f3(x)for x in X]
plt.plot(X,Y3,'m', label='a=2,b=6')

#Titles and labels
plt.legend()
plt.title('Uniform Distribution\nProbability Density Function')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.show()
```

```python
#calculate mean,median and mode

a,b=1,3
mean=(a+b)/2 # In uniform distribution both mean and median are same
median=mean
mode=np.random.randint(low=a,high=b,size=1)
variance=(b-a)**2/12
print('Variance is :',variance)
#plot mean,median and mode
plt.plot(mean,0,"s",label='mean and median')
plt.plot(mode,0,"-o",label='mode')

#Probability Density Function
f=un_pdf(a,b)
Y=[f(x)for x in X]
plt.plot(X,Y,'m', label='a=2')
plt.legend()
plt.show()
```

## Cumulative Distributive Function(CDF)

```python
import numpy as np
import matplotlib.pyplot as plt

#Define CDF in Uniform Distribution

def un_cdf(a,b):
    def fn(x):
        if x<a:
            return 0
        elif a<=x<=b:
            return (x-a)/(b-a)
        elif x>b:
            return 1
    return fn
#Take differnt values for a and b

f1=un_cdf(3,8) #a=3,b=8
f2=un_cdf(1,3) #a=1,b=3
f3=un_cdf(2,6) #a=2,b=6 
# plot functions for values of a taken

#graph1
X=np.linspace(0,10,20)
Y1=[f1(x) for x in X]
plt.plot(X,Y1,'r', label='a=3,b=8')

#graph2
Y2=[f2(x)for x in X]
plt.plot(X,Y2,'y', label='a=1,b=3')

#graph3
Y3=[f3(x)for x in X]
plt.plot(X,Y3,'m', label='a=2,b=6')

#Titles and labels
plt.legend()
plt.title('Uniform Distribution\nCumulative Distributive Function')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.show()
```

# Exponential Distribution

## Probability Density Function(PDF)

```python
import numpy as np
import matplotlib.pyplot as plt

#Define PDF in Exponential Distribution

def exp_pdf(a):
    def fn(x):
        if x<0:
            return 0
        else:
            return a*np.exp((-a)*x)
    return fn

#Take differnt values for a

f1=exp_pdf(0.5) #a=0.5
f2=exp_pdf(1) #a=1
f3=exp_pdf(2) #a=2 
# plot functions for values of a taken

#graph1
X=np.linspace(0,4,20)
Y1=[f1(x) for x in X]
plt.plot(X,Y1,'r', label='a=0.5')

#graph2
Y2=[f2(x)for x in X]
plt.plot(X,Y2,'y', label='a=1')

#graph3
Y3=[f3(x)for x in X]
plt.plot(X,Y3,'m', label='a=2')

#Titles and labels
plt.legend()
plt.title('Exponential Distribution\nProbability Density Function')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.show()
```

```python
#calculate mean and median

a=2
mean=1/a
median=np.log(2)/a
mode=0
variance=1/a**2
print('variance is:',variance)

#plot mean and median
plt.plot(mean,0,"s",label='mean')
plt.plot(median,0,"-o",label='median')
plt.plot(mode,0,"d",label='mode')
#Probability Density Function
f3=exp_pdf(a)
Y3=[f3(x)for x in X]
plt.plot(X,Y3,'m', label='a=2')
plt.legend()
plt.show()
```

## Cumulative distributive Function(CDF)

```python
import numpy as np
import matplotlib.pyplot as plt

#Define CDF in Exponential Distribution

def exp_cdf(a):
    def fn(x):
        if x<0:
            return 0
        else:
            return 1-np.exp((-a)*x)
    return fn
#Take differnt values for a

f1=exp_cdf(0.5) #a=0.5
f2=exp_cdf(1) #a=1
f3=exp_cdf(2) #a=2 
# plot functions for values of a taken

#graph1
X=np.linspace(0,4,20)
Y1=[f1(x) for x in X]
plt.plot(X,Y1,'r', label='a=0.5')

#graph2
Y2=[f2(x)for x in X]
plt.plot(X,Y2,'y', label='a=1')

#graph3
Y3=[f3(x)for x in X]
plt.plot(X,Y3,'m', label='a=2')

#Titles and labels
plt.legend()
plt.title('Exponential Distribution\nCumulative Distributive Function')
plt.xlabel('x')
plt.ylabel('P(X<=x)')
plt.show()
```

# Gamma Distribution

## Probability Density Function(PDF)

```python
import numpy as np
import matplotlib.pyplot as plt
import math

#Define PDF in Gamma Distribution
def gamma_pdf(k,t):
    def fn(x):
        if x>0:
            f=pow(x,k-1)*np.exp(-x/t)/(math.gamma(k)*pow(t,k))
            return f
        else:
            return 0
    return fn    
#Take differnt values for k and t

f1=gamma_pdf(1,2) #k=1,t=2
f2=gamma_pdf(5,1) #k=5,t=1
f3=gamma_pdf(9,0.5) #k=9,t=0.5 
# plot functions for values of a taken

#graph1
X=np.linspace(0,10)
Y1=[f1(x) for x in X]
plt.plot(X,Y1,'r', label='k=1,t=2')

#graph2
Y2=[f2(x)for x in X]
plt.plot(X,Y2,'y', label='k=5,t=1')

#graph3
Y3=[f3(x)for x in X]
plt.plot(X,Y3,'m', label='a=9,b=0.5')

#Titles and labels
plt.legend()
plt.title('Gamma Distribution\nProbability Density Function')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.show()                    
                    
            
```

```python
#calculate mean,median and mode

k,t=5,1
mean=k*t
mode=(k-1)*t # valid for k>=1
variance=k*t**2
print('variance is :',variance)
#plot mean,median and mode
plt.plot(mean,0,"s",label='mean')
plt.plot(mode,0,"-o",label='mode')

#Probability Density Function
f=gamma_pdf(k,t)
Y=[f(x)for x in X]
plt.plot(X,Y,'m', label='k=5,t=1')
plt.legend()
plt.show()
```

## Cumulative Distribution Function(CDF)

```python
# Import libraries
from scipy.stats import gamma
def gamma_cdf(a,b):
    y=gamma.cdf(x,a,loc=0,scale=b)
    return y
x=np.linspace(0,10)
# Take functions for different values of a and b

y1=gamma_cdf(1,2)
y2=gamma_cdf(5,1)
y3=gamma_cdf(9,0.5)

# plot the graphs

plt.plot(x,y1,'r',label='a=1,b=2')
plt.plot(x,y2,'y',label='a=5,b=1')
plt.plot(x,y3,'m',label='a=9,b=0.5')
#Titles and legends

plt.title('cumulative distributiion function')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.legend()
plt.show()
```

# Beta Distribution

## Probability Density Function(PDF)

```python
from scipy.stats import beta
def beta_pdf(a,b):
    y=beta.pdf(x,a,b)
    return y
x=np.linspace(0,1)
# give differnt values of a and b
y1=beta_pdf(0.5,0.5)#a=0.5,b=0.5
y2=beta_pdf(5,1)#a=5,b=3
y3=beta_pdf(2,5)#a=2,b=5
#plot the graphs

plt.plot(x,y1,'r',label='a=0.5,b=0.5')
plt.plot(x,y2,'y',label='a=5,b=3')
plt.plot(x,y3,'m',label='a=2,b=5')
#Titles and legends

plt.title('Beta Distribution\nProbability Density Function')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.legend()
plt.show()
```

```python
#calculate mean,median and mode

a,b=5,1
mean=a/(a+b)
median=(a-(1/3))/(a+b-(2/3)) # valid for a,b>1
mode=(a-1)/(a+b-2)#valid for a,b>1
variance=a*b/(((a+b)**2)*(a+b+1))
print('variance is :',variance)

#plot mean,median and modepl
plt.plot(mean,0,"s",label='mean')
plt.plot(mode,0,"-o",label='mode')
plt.plot(median,0,"v",label='median')

#Probability Density Function
y=beta_pdf(5,1)
plt.plot(x,y,'m', label='k=5,t=1')
plt.legend()
plt.show()
```

## Cumulative Distribution Function(CDF)

```python
# import libraries
import scipy.special as sc
def beta_cdf(a,b):
    y=np.piecewise(x,[(x>=0)&(x<=1)],[lambda x:sc.betainc(a,b,x)])
    return y
x=np.linspace(0,1)
# give differnt values of a and b
y1=beta_cdf(0.5,0.5)#a=0.5,b=0.5
y2=beta_cdf(5,1)#a=5,b=3
y3=beta_cdf(2,5)#a=2,b=5
#plot the graphs

plt.plot(x,y1,'r',label='a=0.5,b=0.5')
plt.plot(x,y2,'y',label='a=5,b=3')
plt.plot(x,y3,'m',label='a=2,b=5')
#Titles and legends

plt.title('Beta Distribution\nCumulative Distribution Function')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.legend()
plt.show()
```

# Gaussian Distribution

## Probability Density Function

```python
import numpy as np
import matplotlib.pyplot as plt
import math

#Define PDF in Gaussian Distribution
def guass_pdf(m,v):
    def fn(x):
        return np.exp(-0.5*pow((x-m)/v,2))/(math.sqrt(pi*2))*v
    return fn    
#Take differnt values for m and v

f1=guass_pdf(0,1) #m=0,v=1
f2=guass_pdf(0,2) #m=0,v=2
f3=guass_pdf(-2,2.5) #m=-2,v=2.5 
# plot functions for values of a taken

#graph1
X=np.linspace(-10,10)
Y1=[f1(x) for x in X]
plt.plot(X,Y1,'r', label='m=0,v=1')

#graph2
Y2=[f2(x)for x in X]
plt.plot(X,Y2,'y', label='m=0,v=2')

#graph3
Y3=[f3(x)for x in X]
plt.plot(X,Y3,'m', label='m=-2,v=2.5')

#Titles and labels
plt.legend()
plt.title('Gaussian distribution\nProbability Density Function')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.show()              
```

## Cumulative Distributive Function(CDF)

```python
import numpy as np
import matplotlib.pyplot as plt
import math

#Define CDF in Gaussian Distribution
def guass_cdf(m,v):
    def fn(x):
        return 0.5*(1+math.erf((x-m)/(math.sqrt(2)*v)))
    return fn
#Take differnt values for m and v

f1=guass_cdf(0,1) #m=0,v=2
f2=guass_cdf(0,2) #m=0,v=5
f3=guass_cdf(-2,2.5) #k=-1,t=3 
# plot functions for values of a taken

#graph1
X=np.linspace(-10,10)
Y1=[f1(x) for x in X]
plt.plot(X,Y1,'r', label='m=0,v=1')

#graph2
Y2=[f2(x)for x in X]
plt.plot(X,Y2,'y', label='m=0,v=2')

#graph3
Y3=[f3(x)for x in X]
plt.plot(X,Y3,'m', label='m=-2,v=2.5')

#Titles and labels
plt.legend()
plt.title('Gaussian distribution\nCumulative Distributive Function')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.show()              
```

# Chi-squared Distribution

## Probability Density Function

```python
import numpy as np
import matplotlib.pyplot as plt
import math

#Define PDF in Chi-squared Distribution
def chi_pdf(k):
    def fn(x):
        if x>0:
            f=pow(x,k*0.5-1)*np.exp(-x/2)/(pow(2,k/2)*math.gamma(k/2))
            return f
        else:
            return 0
    return fn    
#Take differnt values for a and b

f1=chi_pdf(2) #k=2
f2=chi_pdf(5) #k=5
f3=chi_pdf(9) #k=9 
# plot functions for values of a taken

#graph1
X=np.linspace(0,10)
Y1=[f1(x) for x in X]
plt.plot(X,Y1,'r', label='k=2')

#graph2
Y2=[f2(x)for x in X]
plt.plot(X,Y2,'y', label='k=5')

#graph3
Y3=[f3(x)for x in X]
plt.plot(X,Y3,'m', label='k=9')

#Titles and labels
plt.legend()
plt.title('Chi-squared Distribution\nProbability Density Function')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.show()                    
```
